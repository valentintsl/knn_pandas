<h3>Initialisation de la valeur de k</h3>

```python
k = 5
```

<h3>Calculer la distance euclidienne</h3>

```python
import math
def distance_eu(instance1, instance2, taille):
    distance = 0
    for x in range(taille):
        distance += pow((instance1[x]-instance2[x]), 2)
    return math.sqrt(distance)
```

```python
#Exemple
data1 = [2,2,2,'a']
data2 = [4,4,4,'b']
distance = distance_eu(data1,data2, 3)
print ("Distance : ", distance)
```

```python
import numpy as np

def distance_entre_data(data, instancetest):
    df_distances = data.copy()
    liste_data = data.values.tolist()
    df_distances['Distance'] = 0.00
    for i in range(len(df_distances.index)):
        df_distances['Distance'][i] = distance_eu(liste_data[i], instancetest, 4)
        #print(distance_eu(liste_data[i], instancetest, 4))
    return df_distances

```

<h3>Ordonner les distances par ordre croissant</h3>

```python
def ordonner(data, instancetest):
    return distance_entre_data(data,instancetest).sort_values(by = 'Distance', ascending = True)
```

```python
#Exemple
ordonner(df_iris, [3,4,5,6])
```

<h3>Prendre les tops k trouvés</h3>

```python
def top_k(data, instancetest):
    return ordonner(data, instancetest).head(k)
```

```python
top_k(df_iris, [3,4,5,6])
```

<h3>Retourner la classe la plus fréquente</h3>

```python
from collections import Counter
def classe_frequente(data, instancetest):
    compteur = Counter(top_k(data, instancetest)['Class label'])
    return max(compteur, key = compteur.get)
```

```python
#Exemple
classe_frequente(df_iris, [3,4,5,6])
```
